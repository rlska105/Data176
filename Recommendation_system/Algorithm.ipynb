# 추천 시스템 대표적인 모델
1. Prediction Version
  - Matrix Completion Problem
  - 사용자와 아이템 간의 평가 점수를 예측
  - m명의 user, n개의 item이 있을때, m * n matrix를 계산하는 방식<br><br>
2. Ranking Version
  - Top-k Recommendation Problem
  - 특정 유저에게 k개의 적합한 아이템을 추천해주는 방식

# Content-based Recommenders
  : 특정 아이템에 기초하여 비슷한 아이템을 추천해주는 방식
  - 아이템끼리의 유사도를 측정
  - 아이템의 metadata를 사용
    ex) 장르, 감독, 묘사, 배우...
  - 사람들간을 비교하지 않고, 아이템 특성만을 비교
  - 사용자가 적은 초반의 추천 시스템의 경우 대부분 Content-based recommender을 사용
  
  
  1. 장·단점<br>
    1) 장점
      - 다른 사용자의 영향을 받지 않음
      - 새로운 아이템에 대해서도 추천이 가능
      - 추천을 설명하기 쉬움
    
    2) 단점
      - 새로운 유저에게 추천이 불가능
      - 소리, 영상, 이미지 등의 콘텐츠로부터 추천을 위한 metadata, 특징을 추출하기가 어려움
  <br><br>
  2. 동작 로직<br>
    가. 아이템의 분류를 통해 콘텐츠를 분석<br>
    나. 각각의 유저의 선호를 대표하는 유저의 profile을 학습<br>
    다. 아이템의 분류 & 사용자 선호도에 따라 각각의 유저를 위한 추천 리스트를 생성<br>
  <br><br> 
  3. 사용하는 ML 기술<br>
    - KNN Classification(nearest neighbor)<br>
    - Linear Classification<br>
  <br><br> 
  4. 콘텐츠 묘사(전처리?)<br>
    - 구조화된 콘텐츠<br>
    - 구조화되지 않은 콘텐츠<br>
      - ex) keyword<br>
      - 'bag or words'(bow) 방식 또는 'TF/IDF' 방식 사용. 단어의 중요성을 검토하여 사용.<br>
  <br><br>
  5. 아이템 유사도
    - Cosine Similarity
    - 그 외 다른 적합한 유사도

# Collaborative Filtering Recommenders(CF)
  : 사용자 그룹이 형성되어 있고, 그들 간의 평가 점수와 선호도를 고려하여 사용자의 예측 점수와 선호도가 결정됨
    즉, 사용자와 비슷한 다른 사용자를 찾아서 그 사용자는 어떤 평가를 했는지 살펴보는 방식
  
  - 사용자가 어느 정도 형성 되어있고, 데이터가 존재할 때 사용 가능한 추천 알고리즘
  - 종류 : memory-based, model-based, hybrid
  
  1. 장·단점<br>
    1) 장점
      - 어떠한 아이템에 대해서도 추천이 가능하다.(=아이템의 특성에 의존하지 않는다)<br>
      - 별도의 메타데이터가 없어도 잘 동작함
    
    2) 단점
      - 평가되지 않은 아이템에 대해 추천을 하지 않는다.(new-item problem)
      - 보통 가장 인기 있는 아이템을 추천한다
      - 비슷한 유저 군이 존재하는 사용자 그룹이 어느 정도 데이터가 확보되어 있어야 한다.(cold start problem for new users)
      - 데이터가 적은 경우(Cold Start & Data Sparsity)에는 효과를 보여주기 어렵고, Scalability 문제가 있다는 점 그리고 모든 도메인에 적용하기는 어렵다
  <br><br>
  2. 동작 로직<br>
    1) Input<br>
      - 주어진 유저-아이템 간의 평가점수 matrix<br>
      
    2) Output<br>
      - 특정 아이템에 대한 선호도 예측 점수
      - top-N 추천 리스트

# Memory-based CF

  - 전통적인 방법이며 이미 있는 데이터(메모리)를 기반으로 필터링을 하는 것

1. User-Based filtering
  : 유저들의 평가 점수를 바탕으로 예측하고자 하는 유저의 점수를 예측하는 방식

  - 동작 로직<br>
    1) 유저 간의 유사성 계산(Pearson Correlation)<br>
    2) 예측<br>
  <br><br>
  - 장·단점<br>
    1) 장점<br>
      - 구현하기 쉬움
      - 문맥이 독립적
      - contents-based에 비해 정확
    
    2) 단점
      - 희소성 : 평가를 하는 사용자가 적음
      - 확장성 : 사용자 그룹이 커질수록 비용이 비쌈
      - Cold Start : 신규 사용자에게 추천이 힘듦
<br><br>

2. Item-based Filtering
  : 아이템 사이에 유사성을 판단하여 예측(컨텐츠 기반과의 차이점 : KNN Step을 건너뜀)
  
  - 사용자 그룹을 사용하여 관계를 판단하기에 CF에 포함된다
  - user_based보다 더 좋은 성능을 보여 줌
<br><br>

3. Hybrid Recommender Systems
  - 데이터와 상황에 따라 Content-based와 Collaborative Filtering을 적절히 섞어서 사용하는 방식

### User_based VS Item-based
- 아마존과 넷플릭스를 비롯한 서비스에서는 대부분 아이템 기반을 활용한다고 알려져 있다.<br>
  또 적절한 UX를 위해서라면, 유저가 아이템을 평가하는 순간 다른 아이템을 추천할 수 있어야 하는데, 매 평가시마다 유사도 정보를 업데이트하는 것은 현실적으로 어려울 것이다.<br>
  아이템 기반에서는 일정 기간마다 유사도를 업데이트 하는 것으로도 충분히 위와 같은 적절한 UX를 제공할 수 있다.<br><br>

- 대부분의 경우는 사용자에 비해 아이템 수가 적기 때문에, 아이템 간의 관계 데이터가 발생할 확률이 높다.<br>
  따라서 데이터가 누적될수록 추천의 정확도가 높아질 가능성이 더 높다.

# Model-based CF
  - 기계학습을 적용한 필터링 방식으로 사용자(평가 주체)와 아이템(평가 대상)이 숨겨진 특성 값을 계산하여 필터링 하는 것
  -  Memory-based 방식에서는 별도의 모델을 만들지 않았기 때문에 평점을 가지고 있지 않은 것에 대해서는 효과가 없었던 반면<br>
     Model-based 방법에서는 과거의 사용자 평점 데이터를 이용해 모델을 만들었기 때문에 평점 정보가 없다하더라도 특정 아이템에 대한 사용자의 평점을 예측할 수 있다.
  - 의사결정트리, SVM등의 머신러닝 알고리즘을 이용해 평점예측모델을 만든다.<br><br>
  
  - 모델을 만드는데 사용되는 방법<br>
    1) Matrix Factorization<br>
       - SVD(Singular Value Decomposition), PCA(Principal Component Analysis)<br>
    
    2) Association Rule Mining<br>
       - Shopping basket analysis<br>
    
    3) Probabilistic Models<br>
       - Clustering, Bayesian networks<br>
    
    4) Other Techniques<br>
       - Regression, Deep Learning, SVM<br>

# 추천시스템 평가 방법
 - 사용자 평가, 온라인 평가, 오프라인 평가
 - 사용자 평가와 온라인 평가는 사용자에 의한 평가, 가장 큰 차이점은 추천 시스템의 평가를 위해 사용자를 활용하는 시점<br>
   사용자 평가는 특정 시점에 사용자를 초청해 추천시스템의 성능을 점검, 온라인 평가는 추천시스템이 실제 환경에서 활용되고 있는 상황에서 사용자가 보여주는 행위를 보고 성능을 평가하는 방식
 - 오프라인 평가는 과거의 데이터를 이용해 평가하는 방식으로 데이터 기반의 평가
 <br><br>
 
 1. 사용자 평가
   : 사용자를 초청하고 추천시스템을 사용 후 피드백을 받아 추천시스템을 평가
   - 장점 : 초청한 사용자들에게 원하는 정보를 수집할 수 있다(추천시스템 선호도, 사용상의 문제점 등)
   - 사용자가 추천시스템을 평가 하는걸 인지하고 있기 때문에 자신의 의사를 100% 표현할 수 없을 수 있고 환경적인 요소로 왜곡된 설문 결과가 나올 수 있음을 주의해야 한다
 <br><br>
 2. 온라인 평가
   : 사용자를 별도로 초청하지 않고 실제 환경에 추천시스템을 적용 후 사용자의 행위를 보고 성능을 파악
   - 장점 : 사용자가 추천시스템 평가를 하고 있다고 인지하지 않기 때문에 사용자 평가에서 발생하는 문제가 나타나지 않음
   - A/B 테스트를 통해 추천시스템의 성능을 평가
 <br><br>
 3. 오프라인 평가
   : 과거의 데이터를 이용하여 성능 평가
   - 추천시스템을 만들고 해당 시스템에 과거의 데이터를 넣어 얼마나 정확하게 예측하는지를 평가
   - 평가방법과 평가항목을 표준화시킬 수 있기 때문에 추천시스템 평가에서 자주 활용된다
   - 평가항목 : Accuracy, Coverage, Confidence, Novelty 등

### 코사인 유사도 문제점

#### First case<br>

(5, 5, 5, 5, 5, 5), (1, 1, 1, 1, 1, 1).<br>
이 때, 이 둘간의 유사도는 1이다.<br>
두 아이템에 대한 사용자들의 평가는 극명하게 갈리는데, 이 둘의 유사도가 1이라고 한다.<br>
이렇게 코사인 유사도에서는 유저마다의 개인적인 평가 성향을 반영하지 못한다는 단점이 있다.<br>
이를 보완하기 위해, 앞서 말한 피어슨 유사도, 혹은 약간의 보정 과정을 거친 코사인 유사도를 사용할 수도 있다.<br><br>

#### Second case<br>

어떤 두 아이템에 대해서 유사도를 구하려 하는데, 이 두 영화를 모두 평가한 사람이 한명뿐인 경우에 유사도가 역시 1이 나오게 된다<br>
이렇게 적은 데이터에 기반한 정보는 추천의 정확도를 떨어뜨릴 가능성이 높다.<br>
이런 경우를 방지하기 위해 최소 평가 인원의 수를 정하는 것이다.<br>
두 아이템을 모두 평가한 사람의 수가 5명 이상일 때부터 유사도를 계산하기로 하고, 5명 미만의 공통 평가 인원수를 갖는 두 아이템에 대한 유사도를 0으로 처리하는 것이다.
