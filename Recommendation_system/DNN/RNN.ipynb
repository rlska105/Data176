## RNN(Recurrent Neural Network, 순환신경망)
  - 기존 뉴럴 네트워크 알고리즘은 이미지처럼 고정된 크기의 입력을 다루는데 탁월하지만, 가변적인 크기의 데이터를 다루기에 적합하지 않음
  - 이런 시퀀스 데이터를 모델링 하기 위해 RNN이 등장
  - 기존 뉴럴 네트워크와 다른 점은 기억(hidden state)을 가지고 있다
    * 기억 : 지금까지의 입력 데이터를 요약한 정보
  - 새로운 입력이 들어올때마다 기억을 조금씩 수정 > 최종 남겨진 기억은 시퀀스 전체를 요약하는 정보
