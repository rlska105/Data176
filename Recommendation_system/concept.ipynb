# 1. 추천시스템 정의·고려 요소
* 정의 : 특정 시점에 특정 고객이 좋아할 만한 리스트를 찾아주는 것<br><br>
- 추천 시스템은 거의 대부분 B2C(business to customer) 비즈니스 환경에서 적용된다.<br><br>
- 고려 할 사항 : 고객, 아이템, 시스템의 업데이트, 시스템의 알고리즘적 성능
  - 시스템 업데이트 : 얼마나 자주 추천되는 리스트가 업데이트 되는가
  - 알고리즘적 성능<br>
    - 추천 모델의 계산량이나 연산 속도
    - big데이터를 처리할 수 있는 좋은 환경
    - CF(Collaborative Filtering)과 DL은 하둡 환경 등의 빅데이터 시스템이 안정적으로 구축되어 있지 않아 좋은 성능이라고 볼 수는 없다.

# 2. 장애요소
- Sparsity Problem : 추천할 아이템과 고객의 증가에 비해 실제 소비량이 따라가지 못함.(차원의 저주 같은 느낌)
  - 서비스가 성장하는 단계에서 더욱 심하게 나타남
  - 소비의 추세가 오프라인에서 온라인으로 넘어가는 상황에서 대부분의 서비스가 양적 팽창을 겪고있음
  - 추천 시스템의 기술적 발전은 대부분 이 문제를 해결하는 것을 염두해 두고 있다.
<br><br>
- Information Utilization Problem : 아이템의 호감도를 어떻게 측정하냐에 대한 문제
  - 영화맛집 추천 서비스가 아닌 대부분의 추천 서비스에서 나타나는 문제점
  - Implicit Score(암묵 점수) : 구매를 위해 하는 사전 행동같은 느낌
  - Explicit Score(명시 점수) : 영화 평점에 대한 raing같은 점수
    - 항상 좋은 데이터는 아님, 잘 정리된 명시 점수의 경우 Sparsity Problem을 심각하게 겪게된다

# 3. 과거 추천 시스템 동향
  - e-commerce의 경우 크게 3가지 분류의 데이터인 Implicit Score, Explicit Score, Contents(아이템 상세 정보)를 시스템 구축에 활용할 수 있다<br><br>
  - 과거엔 KDD(Knowledge Discover in Database) 방법론에 기반하여 알고리즘 개발을 하였다. 물론, 현재의 추천 시스템 역시 넓은 의미로는 KDD라고 할 수 있지만, 여기서는 좁은 의미의 KDD 방법론을 의미
    - 가장 대표적인 방법으로 Association Rule(AR)<br><br>
  - Association Rule(AR, 장바구니 분석)
    - 기초적인 확률론에 기반한 방법, 상품간의 연관성을 수치화하여 나타냄<br><br>
  - 3가지의 추천 타입<br>
    1) Personalized recommender : 유저의 정보에 기반하여 자동으로 아이템 리스트를 추려줌<br>
    2) Non-personalized recommender : rating기반의 인기상품이나 현재 상품 기준 AR 순위 상품을 추천<br>
      - 주로 Cold Start Problem(개인화 추천 모델링을 위한 유저·아이템 정보가 부족한 상황)이 발생하는 상광이나 개인화 추천이 잘 적용되지 않는 추천 영역에 사용 됨
      
    3) Attribute-based recommender(=Content-based approach) : Contents 정보를 활용하여 추천하는 방법, cold start 문제를 해결하는 조금 더 세련된 방법

# 현재 추천 시스템의 동향
  - 과거와 크게 다르지 않지만 딥러닝을 위시한 부분적으로 고도화된 알고리즘들이 적용되고 있으며, 하둡으로 대표되는 빅데이터 처리 플랫폼이 매우 안정화되었다.<br><br>
  - 현재 추천 모델링에 사용되는 3가지 방법론<br><br>
    1) 데이터에 대한 가장 기본적인 접근법인 통계 기반 모델링<br>
      - 카이제곱 검정 이용, 때때로 강력하지만 대부분 좋지 않음
    
    2) 협업 필터링(CF)<br>
      - 가장 널리 알려져 있으며 대부분의 추천 시스템의 구축 방법<br>
      - CF 모델링을 기본으로 다양한 응용 알고리즘이 있으며 이를 발전시키거나 다른 알고리즘과 hybrid하게 접목하는 방향으로 state of the art가 옮겨가고 있다<br>
      - 일부 논문등의 자료에서는 AR도 CF 기반의 모델링으로 분류해버리기도 함. 그 이유는 연관성이라는 측면 때문
      
    3) 딥러닝 기반의 모델링<br>
      - 비교적 최근의 연구결과를 바탕으로 부분적으로 적용되고 있는 방법<br>
      - Content based 모델링 기법에서 이미지 등의 특별한 데이터를 다루기 위한 모델링을 하기도 하며 word2vec으로 대표되는 언어 처리 알고리즘을 키워드 분류 등에 사용하기도 함<br>
      - 현재 가장 활발하게 연구 및 개발이 이루어지고 있음<br>
    - CF는 취향을 추천해주는 경우 딥러닝은 이미지 추천과 같이 기계가 자동으로 정보를 추출해주어야 하는 경우 성능이 좋을 수 있다    

# 추천 시스템 평가 방법
  - 알고리즘의 목표는 고객의 score와 예측 사이의 오차를 줄이는것이지만, 추천 시스템을 구축하는 사람에게는 다른 평가 기준이 필요함<br><br>
  - B2C 추천 시스템에서 추천 결과를 평가하는 2가지 방법<br>
    1) 오프라인 모델 : 알고리즘의 성능 자체를 평가<br>
      - 일반적인 ML 평가지수 사용(Accuracy, Precision, Recall, F1-score, ROCurve, RMSE, MAPE, nDCG(검색분야) 등)<br>
    2) 온라인 모델 : 알고리즘이 예측한 리스트와 실제 유저들이 좋아하는 리스트를 다양한 기준으로 비교해 보는 실전 테스트
      - 완성된 알고리즘을 유저들에게 배포했을 때 어떤 결과가 나오는지 A/B 테스트 등으로 실험
      - 기술적인 평가 지표인 CR, CTR 등이 들어가지만, 실험이라는 의미가 큼
      - 도메인마다 다른 실험 측정 기준을 가지고 있고, 페이지의 체류시간이나 구매액 등을 평가하기도 함
  - Serendipity
    - 추천 시스템에서 정성적 평가 기준을 부르는 말
    - 상품의 우연성을 나타 냄
    - 인간만이 인지할 수 있는 제 3의 요소, 티셔츠를 구매했을 때 조금 다른 카테고리의 의류를 보여주는 등의 방식
